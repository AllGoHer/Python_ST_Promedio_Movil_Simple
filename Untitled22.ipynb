```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Cargar los datos desde el archivo CSV
```

```python
turismo = pd.read_csv('C:\\Users\\ALLAN\\Desktop\\Business Analitics con Python y ChatGPT\\turistas.csv', index_col = 'Mes')
```

```python
turismo.head()
```

```python
# Asegurarse de que el índice esté en formato de fecha
```

```python
turismo.index = pd.to_datetime(turismo.index)
```

```python
# Calcular el Promedio Móvil Simple (PMS) para 3 meses
# .rolling() es un método que proporciona la capacidad de aplicar funciones a una ventana móvil de datos.
# El argumento window=3 define el tamaño de la ventana móvil. En este caso, estamos utilizando una ventana 
# de 3 períodos (meses). Esto significa que el promedio se calcula utilizando el valor actual y los dos valores anteriores.
```

```python
turismo['PMS3meses'] = turismo['Turistas'].rolling(window=3).mean()
```

```python
# Calcular el Promedio Móvil Simple (PMS) para 6 meses
```

```python
turismo['PMS6meses'] = turismo['Turistas'].rolling(window=6).mean()
```

```python
# Visualización de los datos
```

```python
plt.figure(figsize=(12,6), dpi = 300)
plt.plot(turismo['Turistas'], label = 'Turistas (miles)')
plt.plot(turismo['PMS3meses'], label = 'PMS 3 meses')
plt.plot(turismo['PMS6meses'], label = 'PMS 6 meses')
plt.title('Análisis de Turismo en México con Promedio Móvil Simple')
plt.xlabel('Fecha')
plt.ylabel('Número de Turistas en Miles')
plt.legend()
plt.show()
```
